Figure S24A,B in M & S (2020) A&T (in the SI)

This figure illustrates the detection rate of the targetted contaminants (SVOCs and trace metals) per number of extraction cycles.

Plot type: Bar plot

```python
### Figures created in a loop of two items

from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

for item in ['overall', 'after-sieve']:
    df = pd.read_excel(r'detection_percentages.xlsx', sheet_name= item)
    df = df[df['case'] != 'Best']
    
    for var in [col for col in list(df.columns) if 'numb' in col]:
        df[var] = (df[var]/41) * 100
    
    df['counter'] = df.index + 1
    df.loc[df['case'] == 'Medium', 'counter'] = df['counter'] + 1
    
    
    df.replace({'Trace metals (with As)' : 'TM w As',
                'Trace metals (without As)' : 'TM w/o As'}, inplace = True)
    
    base_width = (1/7)
    
    x = list(df['analyte'])
    x.insert(6, ' ')
    
    plt.figure()
    plt.bar(df['counter'] -0.5, df['number_1'], color = 'k', width = base_width, edgecolor = 'k', linewidth  = 0.5)
    plt.bar(df['counter'] + base_width - 0.5, df['number_2'], color = 'r', width = base_width,  edgecolor = 'k', linewidth  = 0.5)
    plt.bar(df['counter'] + 2 * base_width - 0.5, df['number_3'], color = 'b', width = base_width, edgecolor = 'k', linewidth  = 0.5)
    plt.bar(df['counter'] + 3 * base_width - 0.5, df['number_4'], color = 'green', width = base_width, edgecolor = 'k', linewidth  = 0.5)
    plt.bar(df['counter'] + 4 * base_width - 0.5, df['number_5'], color = 'orange', width = base_width, edgecolor = 'k', linewidth  = 0.5)
    plt.bar(df['counter'] + 5 * base_width - 0.5, df['number_6'], color = 'grey', width = base_width, edgecolor = 'k', linewidth  = 0.5)
    
    
    
    df.replace({'Trace metals (with As)' : 'TM w As',
                'Trace metals (with As)' : 'TM w/o As'}, inplace = True)
    
    for i in range(13):
        plt.text(i+0.9, -2, x[i], fontsize = 5, rotation = 90, va = 'top', ha = 'center')
    
    plt.text(3.3, -20, 'Worst-case Scenario', fontsize = 10, ha = 'center')
    plt.text(10, -20, 'Medium-case Scenario', fontsize = 10, ha = 'center')
        
    plt.xlim(0.3, 13.5)
    plt.xticks([])
    
    plt.ylim(0, 105)
    plt.yticks(np.arange(0, 120, step=20))
    plt.ylabel('Detection Rate (%)', size = 14)
    
    plt.tick_params(axis = 'both', direction = 'inout')
    
    plt.savefig(r'C:/Career/Learning/Python Practice/Stata_Python_Booster/PhD - Extraction/Processed/plots/FigureS24_DR_%s.jpg' %item, format='jpg', dpi=1600, bbox_inches='tight')
```
